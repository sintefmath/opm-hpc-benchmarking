```python
dataorigin="joseki_franzf_14:22:38-02-03-2020/"
maxproc=12

i=0
for casename in ["NORNE_ATW2013", "SPE9"]:
    i+=1
    pl.figure(i)
    pl.clf() 
    pl.title('total time for case '+casename) 
    pl.xlabel('N = num processors') 
    pl.ylabel('t1/(N*tN) *100 %')

    for numthreads in [1, 2]:
        if numthreads == 1:
                mark='o'
        else:
                mark='x'
        res=getTotalTime(dataorigin,casename,numthreads,maxproc)
        pl.plot(ran,100*res[0]/(ran*res),'-'+mark,label=dataorigin+" "+str(numthreads)+" threads")
    pl.legend()
```
