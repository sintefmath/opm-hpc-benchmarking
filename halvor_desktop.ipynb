```python
maxproc=20
numthreadsvec = [1,2]
cases = ["NORNE_ATW2013","SPE9"]
ran=np.arange(1,maxproc+1)
```

```python
dataorigin="hnil-workstation_hnil_09:19:43-26-03-2020/"


i=0
for casename in cases:
    i+=1
    pl.figure(i,figsize=(12,8))
    pl.clf() 
    pl.title('total time for case '+casename) 
    pl.xlabel('N = num processors') 
    pl.ylabel('t1/(N*tN) *100 %')

    for numthreads in numthreadsvec:
        print(numthreads)
        if numthreads == 1:
                mark='o'
        else:
                mark='x'
        res=getTotalTime(dataorigin,casename,numthreads,maxproc)
        pl.plot(ran,100*res[0]/(ran*res),'-'+mark,linewidth=4,markersize=12,label=dataorigin+" "+str(numthreads)+" threads")
    pl.legend()
```

```python
i=0
numthreads=2
for casename in cases:
    i+=1
    pl.figure(i,figsize=(8,8))
    pl.clf() 
    pl.title('Speedup comparison '+casename) 
    pl.xlabel('N = num processors') 
    pl.ylabel('speedup')

    numthreads=1
    overlapcell=getNumOverlapCells(dataorigin,casename,numthreads,maxproc)
    ownedcell=getNumOwnedCells(dataorigin,casename,numthreads,maxproc)
    totalcell=getNumTotalCells(dataorigin,casename,numthreads,maxproc)
    tmp=np.ones(maxproc)
    for j in np.arange(1,maxproc):
        largestpartition = np.max(totalcell[j,:])
        tmp[j] = totalcell[0,0]/largestpartition
    pl.plot(ran, tmp,'ok-',linewidth=4,markersize=12,label="maximum first order partition speedup")
    pl.plot(ran, ran,'--',linewidth=4,markersize=12,label="optimal")
    pl.legend()
```
