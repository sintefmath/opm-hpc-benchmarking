```python
i=0
numthreads=2
for casename in ["NORNE_ATW2013", "SPE9"]:
    i+=1
    pl.figure(i,figsize=(8,8))
    pl.clf() 
    pl.title('Speedup comparison '+casename) 
    pl.xlabel('N = num processors') 
    pl.ylabel('speedup')
    for dataorigin in ["joseki_franzf_14:55:53-06-03-2020/", "login-1.saga.sigma2.no_franzf_09:50:04-11-03-2020/"]:
        if dataorigin=="joseki_franzf_14:55:53-06-03-2020/":
            maxproc=12
            style='og'
            lab='joseki'
        else:
            maxproc=16
            style='xb'
            lab='saga'

        numthreads=1
        overlapcell=getNumOverlapCells(dataorigin,casename,numthreads,maxproc)
        ownedcell=getNumOwnedCells(dataorigin,casename,numthreads,maxproc)
        totalcell=getNumTotalCells(dataorigin,casename,numthreads,maxproc)
        tmp=np.ones(maxproc)
        ran=np.arange(1,maxproc+1)
        for j in np.arange(1,maxproc):
            largestpartition = np.max(totalcell[j,:])
            tmp[j] = totalcell[0,0]/largestpartition
        pl.plot(ran, tmp,style+'-',linewidth=4,markersize=12,label="maximum first order partition speedup "+lab)
        pl.plot(ran, ran,style+'--',linewidth=4,markersize=12,label="optimal "+lab)
        pl.legend(fontsize=14)
```

```python
i=0
numthreads=2
for casename in ["NORNE_ATW2013", "SPE9"]:
    i+=1
    pl.figure(i,figsize=(12,8))
    pl.clf() 
    pl.title('Speedup comparison '+casename) 
    pl.xlabel('N = num processors') 
    pl.ylabel('speedup')
    for dataorigin in ["joseki_franzf_14:55:53-06-03-2020/", "login-1.saga.sigma2.no_franzf_09:50:04-11-03-2020/"]:
        if dataorigin=="joseki_franzf_14:55:53-06-03-2020/":
            maxproc=12
            style='o-'
            lab='joseki'
        else:
            maxproc=16
            style='x-'
            lab='saga'
            
        numthreads=1
        overlapcell=getNumOverlapCells(dataorigin,casename,numthreads,maxproc)
        ownedcell=getNumOwnedCells(dataorigin,casename,numthreads,maxproc)
        totalcell=getNumTotalCells(dataorigin,casename,numthreads,maxproc)
        tmp=np.ones(maxproc)
        ran=np.arange(1,maxproc+1)
        for j in np.arange(1,maxproc):
            largestpartition = np.max(totalcell[j,:])
            tmp[j] = totalcell[0,0]/largestpartition
        pl.plot(ran, tmp,style,linewidth=4,markersize=12,label="maximum first order partition speedup "+lab)

        res=getLinearSolveTime(dataorigin,casename,numthreads,maxproc)
        pl.plot(ran,res[0]/(res),style,linewidth=4,markersize=12,label="actual speedup linear solve time "+lab)

        res=getAssemblyTime(dataorigin,casename,numthreads,maxproc)
        pl.plot(ran,res[0]/(res),style,linewidth=4,markersize=12,label="actual speedup assembly time "+lab)

        totaltime=getTotalTime(dataorigin,casename,numthreads,maxproc)
        pl.plot(ran,totaltime[0]/(totaltime),style,linewidth=4,markersize=12,label="actual speedup total time "+lab)
        pl.legend(fontsize=14)
```
